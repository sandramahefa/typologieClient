# Comportement achat coupon des clients 2021


```python
#Par rapport aux nombre de coupon

def analyseParRapportNbCoupon(datas):
    nbAchatParClient = datas[["N° Fiche client","Achat euro"]].groupby(["N° Fiche client"]).count()
    nbAchat = datas[["N° Fiche client","Achat euro","Cat coupon"]].groupby(["N° Fiche client","Cat coupon"]).count()
    #tAchat = datas[["N° Fiche client","Achat euro","Cat coupon"]].groupby(["N° Fiche client","Cat coupon"]).sum()
    
    ficheClient = []
    catCoupon = []
    nbachatParCat = []
    for i in range(len(nbAchat)):
        ficheClient.append(nbAchat.index[i][0])
        catCoupon.append(nbAchat.index[i][1])
        nbachatParCat.append(nbAchat["Achat euro"][i])
    #liste pour le total broupby client
    totalGroupByClient = []
    for i in range(len(nbAchat)):
        totalGroupByClient.append(nbAchatParClient["Achat euro"][nbAchat.index[i][0]])
    resultatAnalyse = pd.DataFrame({"N° Fiche client":ficheClient,"Cat coupon":catCoupon,"nombre achat":nbachatParCat,"total achat client":totalGroupByClient})
    
    resultatAnalyse["% par catégorie coupon"] = (resultatAnalyse["nombre achat"]/resultatAnalyse["total achat client"])*100

    resultatAnalyse["classe"] = resultatAnalyse["% par catégorie coupon"]
    
    for i in range(len(resultatAnalyse["classe"])):
        if resultatAnalyse["classe"][i] <20:
            resultatAnalyse["classe"][i] = "[0%-20%["
        elif 20<=resultatAnalyse["classe"][i] <40:
            resultatAnalyse["classe"][i] = "[20%-40%["
        elif 40<=resultatAnalyse["classe"][i] <60:
            resultatAnalyse["classe"][i] = "[40%-60%["
        elif 60<=resultatAnalyse["classe"][i] <80:
            resultatAnalyse["classe"][i] = "[60%-80%["
        else:
            resultatAnalyse["classe"][i] = "[80%-100%]"
        
    r = resultatAnalyse[["Cat coupon","nombre achat","classe"]].groupby(["Cat coupon","classe"]).sum()
    #r.pivot_table(index="Cat coupon",columns="classe",values="Achat euro")
    pourcentage_vs_nbCoupon = (r/resultatAnalyse["nombre achat"].sum())*100

    #pourcentage_vs_nbCoupon.pivot_table(index="Cat coupon",columns="classe",values="nombre achat").plot.bar(figsize=(12,4),xlabel="Catégorie coupon",title="pourcentage par rapport au nombre de coupon")
#    print(round(pourcentage_vs_nbCoupon.pivot_table(index="Cat coupon",columns="classe",values="nombre achat"),1))
    return pourcentage_vs_nbCoupon.pivot_table(index="Cat coupon",columns="classe",values="nombre achat")
```

```python
#Par rapport aux CA

def analyseParRapportCA(datas):
    achatParClient = datas[["N° Fiche client","Achat euro"]].groupby(["N° Fiche client"]).sum() #chiffre d'affaire par client
    #print("CA par client", achatParClient)
    #chiffre d'affaire par client et cat coupon
    valeurAchat = datas[["N° Fiche client","Achat euro","Cat coupon"]].groupby(["N° Fiche client","Cat coupon"]).sum()  
    #print("CA par client et catégorie coupon", valeurAchat)
    
   # tAchat = datas[["N° Fiche client","Achat euro","Cat coupon"]].groupby(["N° Fiche client","Cat coupon"]).sum()
    
    ficheClient = []
    catCoupon = []
    CAParCat = []
    for i in range(len(valeurAchat)):
        ficheClient.append(valeurAchat.index[i][0])
        catCoupon.append(valeurAchat.index[i][1])
        CAParCat.append(valeurAchat["Achat euro"][i])
    #liste pour le total broupby client
    totalGroupByClient = []
    for i in range(len(valeurAchat)):
        totalGroupByClient.append(achatParClient["Achat euro"][valeurAchat.index[i][0]])
    resultatAnalyse = pd.DataFrame({"N° Fiche client":ficheClient,"Cat coupon":catCoupon,"Chiffre d'affaire":CAParCat,"total achat client":totalGroupByClient})
    
    resultatAnalyse["% par catégorie coupon"] = (resultatAnalyse["Chiffre d'affaire"]/resultatAnalyse["total achat client"])*100
   
    resultatAnalyse["classe"] = resultatAnalyse["% par catégorie coupon"]
    
    for i in range(len(resultatAnalyse["classe"])):
        if resultatAnalyse["classe"][i] <20:
            resultatAnalyse["classe"][i] = "[0%-20%["
        elif 20<=resultatAnalyse["classe"][i] <40:
            resultatAnalyse["classe"][i] = "[20%-40%["
        elif 40<=resultatAnalyse["classe"][i] <60:
            resultatAnalyse["classe"][i] = "[40%-60%["
        elif 60<=resultatAnalyse["classe"][i] <80:
            resultatAnalyse["classe"][i] = "[60%-80%["
        else:
            resultatAnalyse["classe"][i] = "[80%-100%]"
        
    r = resultatAnalyse[["Cat coupon","Chiffre d'affaire","classe"]].groupby(["Cat coupon","classe"]).sum()
    #r.pivot_table(index="Cat coupon",columns="classe",values="Achat euro")
    pourcentage_vs_CA = (r/resultatAnalyse["Chiffre d'affaire"].sum())*100

    #pourcentage_vs_CA.pivot_table(index="Cat coupon",columns="classe",values="Chiffre d'affaire").plot.bar(figsize=(12,4),xlabel="Catégorie coupon",title="pourcentage par rapport aux CA")
    #print(round(pourcentage_vs_CA.pivot_table(index="Cat coupon",columns="classe",values="Chiffre d'affaire"),1))
    return pourcentage_vs_CA.pivot_table(index="Cat coupon",columns="classe",values="Chiffre d'affaire")
```
